```python
# ==================================================================
# ì…€ 1: CSV ë¬¸ì„œ ë³€í™˜ ë° FAISS ë²¡í„° DB ìƒì„±
# ==================================================================
import os
from pathlib import Path
from typing import List

import pandas as pd
from langchain.docstore.document import Document
from langchain_upstage import UpstageEmbeddings
from langchain_community.vectorstores import FAISS

# --- 0. í™˜ê²½ì„¤ì • ---
os.environ.setdefault("UPSTAGE_API_KEY", "up_EoF0I0CzeHxuDYmf0we544GMPCFIT")
UPSTAGE_API_KEY = os.environ.get("UPSTAGE_API_KEY", "")
if not UPSTAGE_API_KEY:
    raise ValueError("UPSTAGE_API_KEY í™˜ê²½ ë³€ìˆ˜ë¥¼ ì„¤ì •í•˜ì„¸ìš”.")

CURRENT_DIR = Path(__file__).parent if "__file__" in globals() else Path.cwd()
if Path.cwd() != CURRENT_DIR:
    os.chdir(CURRENT_DIR)

VECTOR_DB_DIR = Path("table_kb")
CSV_ROOT = Path.cwd()

embeddings = UpstageEmbeddings(
    api_key=UPSTAGE_API_KEY,
    model="solar-embedding-1-large"
)
print("âœ… Upstage ì„ë² ë”© ëª¨ë¸ ë¡œë“œ ì™„ë£Œ.")


def coalesce(*values: object) -> str:
    for value in values:
        if value is None:
            continue
        if isinstance(value, float) and pd.isna(value):
            continue
        text = str(value).strip()
        if text:
            return text
    return ""


def build_quota_text(row: pd.Series) -> str:
    year = coalesce(row.get("í•™ë…„ë„"), "2019í•™ë…„ë„")
    college = coalesce(row.get("ëŒ€í•™"))
    faculty = coalesce(row.get("í•™ë¶€"))
    major = coalesce(row.get("í•™ê³¼_ì „ê³µ"), row.get("í•™ê³¼ ë˜ëŠ” ì „ê³µ"))
    quota = coalesce(row.get("ì •ì›"), row.get("ì…í•™ì •ì›_ëª…")) or "ë¯¸ìƒ"

    prefix_parts = [year]
    if college:
        prefix_parts.append(college)
    if faculty:
        prefix_parts.append(faculty)
    prefix = " ".join(prefix_parts)
    return f"{prefix} ì†Œì† {major}ì˜ ì…í•™ ì •ì›ì€ {quota}ëª…ì…ë‹ˆë‹¤."


def build_degree_text(row: pd.Series) -> str:
    college = coalesce(row.get("ì„¤ì¹˜ëŒ€í•™"))
    major = coalesce(row.get("í•™ê³¼_ì „ê³µ"))
    degree = coalesce(row.get("í•™ìœ„_ì¢…ë¥˜"))
    quota = coalesce(row.get("ì…í•™ì •ì›_ëª…"))
    period = coalesce(row.get("ì„¤ì¹˜_ìš´ì˜ê¸°ê°„"))
    sentence = f"{college} ì†Œì† {major} ì „ê³µìì—ê²Œ ìˆ˜ì—¬í•˜ëŠ” í•™ìœ„ëŠ” {degree}ì…ë‹ˆë‹¤."
    extras = []
    if quota:
        extras.append(f"ì…í•™ ì •ì›ì€ {quota}ëª…")
    if period:
        extras.append(f"ì„¤ì¹˜Â·ìš´ì˜ ê¸°ê°„ì€ {period}")
    if extras:
        sentence += " " + ", ".join(extras) + "ì…ë‹ˆë‹¤."
    return sentence


def build_contract_text(row: pd.Series) -> str:
    college = coalesce(row.get("ì„¤ì¹˜ëŒ€í•™"))
    form = coalesce(row.get("ì„¤ì¹˜í˜•íƒœ"))
    major = coalesce(row.get("í•™ê³¼_ì „ê³µ"))
    degree = coalesce(row.get("í•™ìœ„_ì¢…ë¥˜"))
    quota = coalesce(row.get("ì…í•™ì •ì›_ëª…"))
    period = coalesce(row.get("ì„¤ì¹˜_ìš´ì˜ê¸°ê°„"))
    parts = [f"ê³„ì•½í•™ê³¼ ì„¤ì¹˜Â·ìš´ì˜ ì •ë³´: ì„¤ì¹˜ëŒ€í•™={college}"]
    if form:
        parts.append(f"ì„¤ì¹˜í˜•íƒœ={form}")
    if major:
        parts.append(f"í•™ê³¼/ì „ê³µ={major}")
    if degree:
        parts.append(f"ìˆ˜ì—¬ í•™ìœ„={degree}")
    if quota:
        parts.append(f"ì…í•™ ì •ì›={quota}ëª…")
    if period:
        parts.append(f"ì„¤ì¹˜Â·ìš´ì˜ ê¸°ê°„={period}")
    return ", ".join(parts) + "."


CSV_CONFIG = [
    {
        "path": CSV_ROOT / "2019_quota.csv",
        "required": ["ëŒ€í•™", "í•™ê³¼_ì „ê³µ"],
        "metadata": {"source": "[ë³„í‘œ 1] 2019í•™ë…„ë„ ì…í•™ì •ì›", "page": 39},
        "builder": build_quota_text,
    },
    {
        "path": CSV_ROOT / "degrees.csv",
        "required": ["ì„¤ì¹˜ëŒ€í•™", "í•™ê³¼_ì „ê³µ", "í•™ìœ„_ì¢…ë¥˜"],
        "metadata": {"source": "[ë³„í‘œ 2] í•™ì‚¬í•™ìœ„ì˜ ì¢…ë¥˜", "page": 51},
        "builder": build_degree_text,
    },
    {
        "path": CSV_ROOT / "contract_dept.csv",
        "required": ["ì„¤ì¹˜ëŒ€í•™", "í•™ê³¼_ì „ê³µ", "í•™ìœ„_ì¢…ë¥˜"],
        "metadata": {"source": "[ë³„í‘œ 3] ê³„ì•½í•™ê³¼ ì„¤ì¹˜Â·ìš´ì˜", "page": 53},
        "builder": build_contract_text,
    },
]


def build_documents(configs: List[dict]) -> List[Document]:
    documents: List[Document] = []
    for cfg in configs:
        csv_path: Path = cfg["path"]
        if not csv_path.exists():
            print(f"âš ï¸ {csv_path.name} íŒŒì¼ì´ ì—†ìŠµë‹ˆë‹¤.")
            continue

        df = pd.read_csv(csv_path)
        missing_cols = [col for col in cfg["required"] if col not in df.columns]
        if missing_cols:
            print(f"âŒ {csv_path.name} ì»¬ëŸ¼ í™•ì¸ í•„ìš”: {missing_cols}")
            continue

        for _, row in df.iterrows():
            text = cfg["builder"](row)
            documents.append(Document(page_content=text, metadata=cfg["metadata"]))

        print(f"ğŸ“„ {csv_path.name} ì²˜ë¦¬ ì™„ë£Œ (ë¬¸ì„œ {len(df)}ê±´).")

    valid_documents = [doc for doc in documents if doc.page_content and doc.page_content.strip()]
    print(f"ğŸ§¹ í•„í„°ë§ í›„ ìœ íš¨ ë¬¸ì„œ ìˆ˜: {len(valid_documents)}")
    return valid_documents


vector_store = None
if VECTOR_DB_DIR.exists():
    try:
        vector_store = FAISS.load_local(
            folder_path=str(VECTOR_DB_DIR),
            embeddings=embeddings,
            allow_dangerous_deserialization=True,
        )
        print("ğŸ“‚ ê¸°ì¡´ ë²¡í„° DB ë¡œë“œ ì™„ë£Œ.")
    except Exception as exc:
        print(f"âš ï¸ ê¸°ì¡´ ë²¡í„° DB ë¡œë“œ ì‹¤íŒ¨: {exc}. ìƒˆë¡œ ìƒì„±í•©ë‹ˆë‹¤.")

if vector_store is None:
    all_documents = build_documents(CSV_CONFIG)
    if all_documents:
        print(f"\nğŸ“¦ ì´ {len(all_documents)}ê°œì˜ ë¬¸ì„œë¥¼ ë²¡í„°í™” ì¤‘ì…ë‹ˆë‹¤...")
        vector_store = FAISS.from_documents(all_documents, embeddings)
        vector_store.save_local(str(VECTOR_DB_DIR))
        print("âœ… ë²¡í„° DB ìƒì„± ì™„ë£Œ (table_kb ì €ì¥).")
    else:
        print("âš ï¸ ì²˜ë¦¬í•  ë¬¸ì„œê°€ ì—†ì–´ ë²¡í„° DBë¥¼ ìƒì„±í•˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤.")
```

```python
# =====================================================================
# ì…€ 3: RAG íŒŒì´í”„ë¼ì¸ êµ¬ì„± ë° ì‹¤í–‰ (solar-pro2 + FAISS)
# =====================================================================

import os
from pathlib import Path
from typing import List

import pandas as pd
from langchain_core.prompts import PromptTemplate
from langchain_upstage import ChatUpstage

# --- 0. í™˜ê²½ì„¤ì • ---
UPSTAGE_API_KEY = os.environ["UPSTAGE_API_KEY"]
CURRENT_DIR = Path(__file__).parent if "__file__" in globals() else Path.cwd()
TESTSET_PATH = CURRENT_DIR.parent / "testset.csv"
TOP_K = 3

if "vector_store" not in globals() or vector_store is None:
    raise RuntimeError("vector_storeê°€ ìƒì„±ë˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤. ë¨¼ì € ë²¡í„°í™” ì…€ì„ ì‹¤í–‰í•˜ì„¸ìš”.")

retriever = vector_store.as_retriever(search_kwargs={"k": TOP_K})

# --- 1. RAG ëª¨ë¸ ë¡œë“œ ë° ì²´ì¸ ì •ì˜ ---
llm = ChatUpstage(api_key=UPSTAGE_API_KEY, model="solar-pro2")

prompt_template = PromptTemplate.from_template(
    """
    Answer ONLY with the final answer.
    If the answer is not present in the context, reply exactly with "The information is not present in the context."
    Do not explain your reasoning.
    ---
    Question: {question}
    ---
    Context:
    {context}
    """
)
chain = prompt_template | llm

print(f"âœ… RAG ì²´ì¸ êµ¬ì„± ì™„ë£Œ (Model: solar-pro2, Top-k: {TOP_K}).")

# --- 2. ì§ˆë¬¸ ë°ì´í„° ë¡œë“œ ---
def read_data(file_path: Path):
    data = pd.read_csv(file_path)
    return data["prompts"], data.get("answers")

try:
    prompts, answers = read_data(TESTSET_PATH)
    print(f"âœ… í…ŒìŠ¤íŠ¸ í”„ë¡¬í”„íŠ¸ {len(prompts)}ê°œ ë¡œë“œ ì™„ë£Œ.")
except FileNotFoundError:
    print(f"âŒ {TESTSET_PATH} íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ê²½ë¡œë¥¼ í™•ì¸í•˜ì„¸ìš”.")
    prompts, answers = [], None

# --- 3. RAG ì‹¤í–‰ ë° ì‘ë‹µ ìƒì„± ---
responses: List[dict] = []
if prompts is not None and len(prompts) > 0:
    print("\nâ–¶ RAG ì‘ë‹µ ìƒì„± ì‹œì‘...")
    #for prompt in prompts:
    for prompt in prompts:
        source_docs = retriever.invoke(prompt)
        context_text = "\n\n".join(doc.page_content for doc in source_docs) or ""
        response = chain.invoke({"question": prompt, "context": context_text})
        responses.append(
            {
                "question": prompt,
                "response": response.content,
                "sources": [doc.metadata for doc in source_docs],
            }
        )
        print(f"\n--- ì§ˆë¬¸: {prompt[:50]}...")
        print(f"--- ì‘ë‹µ: {response.content[:80]}...")
        if not context_text:
            print("--- âš ï¸ ê²€ìƒ‰ëœ ë¬¸ì„œê°€ ì—†ìŠµë‹ˆë‹¤. CSVì™€ ë²¡í„° DBë¥¼ í™•ì¸í•˜ì„¸ìš”.")

    print("\nâœ… RAG ì‘ë‹µ ìƒì„± ì™„ë£Œ.")

    output_path = CURRENT_DIR / "rag_responses.csv"
    pd.DataFrame(responses).to_csv(output_path, index=False)
    print(f"ğŸ’¾ ì‘ë‹µì„ {output_path}ì— ì €ì¥í–ˆìŠµë‹ˆë‹¤.")
else:
    print("âš ï¸ ì‹¤í–‰í•  í”„ë¡¬í”„íŠ¸ê°€ ì—†ìŠµë‹ˆë‹¤.")
```
